```python
import pandas as pd
```

```python
# Read csv
weather_data="weather.csv"
weather_data = pd.read_csv(weather_data)
weather_data
```

```python
# Drop city_name, lat, lon, NaN values, weather_id, weather_icon
# weather_data.drop(columns = ['city_name', 'lat','lon','weather_id','weather_icon'])

weather_data = weather_data.dropna(axis='columns')
weather_data
```

```python
# Filter 10-07-2017 to 10-07-2020
weather_data = weather_data[(weather_data['dt'] >= 1499644800) & (weather_data['dt'] <= 1594339200)]
weather_data
```

```python
#jsonify
from pandas import DataFrame
weather_data.to_json('weather_data.json')
```
